```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
```

```python
def get_mae(max_leaf,x_train,x_test,y_train,y_test):
    model=DecisionTreeRegressor(max_leaf_nodes=max_leaf,random_state=0)
    model.fit(x_train,y_train)
    pred=model.predict(x_test)
    mae=mean_absolute_error(pred,y_test)
    return mae
```

```python
leafs=range(60000,60500,250)
for leaf in leafs:
    mae={}
    mae[leaf]=get_mae(leaf,x_train,x_test,y_train,y_test)
    print('Max_leaf_node : {0} \t\t Mean_absolute_error : {1}'.format(leaf,mae[leaf].round(2)))
    
print('\tBest_leaf_nodes :',min(mae))
```
